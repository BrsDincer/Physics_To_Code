# Position, Displacement, and Average Velocity

## Position

* In physics, position refers to the location of an object relative to a reference point. It is often denoted by the symbol $x$. To describe the motion of an object, you must first be able to describe its position (x), where it is at any particular time. More precisely, we need to specify its position relative to a convenient frame of reference.
* Earth is often used as a frame of reference, and we often describe the position of an object as it relates to
stationary objects on Earth.

## Displacement

* Displacement ($\Delta x$) is a vector quantity that represents the change in position of an object. If an object moves relative to a frame of reference then the objectâ€™s position changes.
* The word **displacement** implies that an object has moved, or has been displaced. Although position is the numerical value of $x$ along a straight line where an object might be located, displacement gives the change in position along this line.
* Since displacement indicates direction, it is a vector and can be either positive or negative, depending on the choice of positive direction. Also, an analysis of motion can have many displacements embedded in it.
* Mathematically, displacement is the final position minus the initial position:
> $\Delta x = x_{final}-x_{initial}$

* The SI unit for displacement is the meter, but sometimes we use
kilometers or other units of length.
* Objects in motion can also have a series of displacements. We define total displacement $\Delta x_{Total}$, as the sum of the individual displacements:
> $\Delta x = \sum_{i=1}^{n} \Delta x_i$.

* The magnitude of the total displacement should not be confused with the distance traveled. Distance traveled, $X_{Total}$, is the total length of the path traveled between two positions. The distance traveled is the sum of the magnitudes of the individual displacements.

## Average Velocity

* Average velocity ($\bar{v}$) is defined as the ratio of the displacement to the total time taken. The time variable allows us not only to state where the object is (its position) during its motion, but also how fast it is moving. How fast an object is moving is given by the rate at which the position changes with time.
* For each position $x_i$, we assign a particular time $t_i$.
* If the details of the motion at each instant are not important, the rate is usually expressed as the average velocity. This vector quantity is simply the total displacement between two points divided by the time taken to travel between them. The average velocity is a vector and can be negative, depending on positions.
* The time taken to travel between two points is called the **elapsed time** ($\Delta t$).
* Average Velocity:
> $\bar{v} = \frac{\Delta x}{\Delta t}$ = $\bar{v} = \frac{x_{2}-x_{1}}{t_{2}-t_{1}}$

# Instantaneous Velocity and Speed

* The quantity that tells us how fast an object is moving anywhere along its path is the **instantaneous velocity**, usually called simply **velocity**.
* It is the average velocity between two points on the path in the limit that the time (and therefore the displacement) between the two points approaches zero.
* The formula for instantaneous velocity is given by:
> $v = \lim_{{\Delta t \to 0}} \frac{\Delta x}{\Delta t}$

* Like average velocity, instantaneous velocity is a vector with dimension of length per time. The instantaneous velocity at a specific time point $t_{0}$ is the rate of change of the position function, which is the slope of the position function $x_{t}$ at $t_{0}$.

* We can calculate the instantaneous speed from the magnitude of the instantaneous velocity:
> $s = \lvert v \rvert$

* When calculating instantaneous velocity, we need to specify the explicit form of the position function $x_{t}$. If each term in the $x_{t}$ equation has the form of $At^{n}$ where $A$ is a constant and $n$ is an integer, this can be differentiated using the power rule to be:
> $\frac{d(At^{n})}{dt} = Ant^{n-1}$

```python
# Given time and position values
timeValues = [0,1,2,3,4,5]
positionValues = [0,2,4,8,12,16]
```

```python
def InstantaneousVelocity(positionValue:int or float,timeValue:int or float,index:int)->int or float:
  deltaX = positionValue[index]-positionValue[index-1]
  deltaT = timeValue[index] - timeValue[index-1]
  return deltaX/deltaT
```

```python
timeIndex = 3
```

```python
velocityAtTime = InstantaneousVelocity(positionValues,timeValues,timeIndex)
print(f"At time t = {timeValues[3]}: Instantaneous Velocity = {velocityAtTime} units per time")
```

```python
def PositionFunction(t:int or float,A:int or float,n:int or float)->int or float:
  return A*t**n
```

```python
def DerivativePowerRule(A:int or float,n:int or float,t:int or float)->int or float:
  return n*A*t**(n-1)
```

```python
def InstantaneousVelocityPowerRule(A:int or float,n:int or float,t:int or float)->int or float:
  return DerivativePowerRule(A,n,t)
```

```python
time = symbols("t")
print(time)
```

```python
positionFunction = 3.0*time+0.5*time**3
print(f"Position Function: {positionFunction}")
```

```python
velocityFunction = diff(positionFunction,time)
print(f"Velocity Function: {velocityFunction}")
```

```python
# Substitute t = 2 into the velocity function to find the instantaneous velocity
instantaneousVelocityAtTime = velocityFunction.subs(time,2)
print(f"Instantaneous Velocity at t = 2 seconds: {round(instantaneousVelocityAtTime.evalf(),2)} m/s")
```

```python
# Function to evaluate the position and velocity functions numerically
evaluatePosition = np.vectorize(lambda x: positionFunction.subs(time,x).evalf())
evaluateVelocity = np.vectorize(lambda x: velocityFunction.subs(time,x).evalf())
```

```python
# Data points for the graph
timeValues = np.linspace(0,4,100) # sec
positionValues = evaluatePosition(timeValues)
```

```python
# Calculate the tangent line at t = 2 seconds
tangentLine = lambda x: evaluateVelocity(2)*(x-2)+evaluatePosition(2)
```

```python
plt.figure(figsize=(11,5))
plt.title("Position vs Time & Tangent (Slope) Line: Velocity At t=2")
plt.plot(timeValues,positionValues,label="Position Function $x(t)$")
plt.plot(timeValues,tangentLine(timeValues),linestyle="--",color="white",label="Tangent Line at $t = 2$")
plt.scatter([2],[evaluatePosition(2)],color="red",marker="o",label="Point $t = 2$",s=130)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Average and Instantaneous Acceleration

* Average acceleration is the rate at which velocity changes. It is calculated as the change in velocity divided by the change in time. The formula for average acceleration is:
> $a_{\text{avg}} = \frac{\Delta v}{\Delta t}$

* Because acceleration is velocity in meters per second divided by time in seconds, the SI units for acceleration are often abbreviated $\frac{m}{s^2}$. This literally means by how many meters per second the velocity changes every second. Recall that velocity is a vector, it has both magnitude and directionâ€”which means that a change in velocity can be a change in magnitude (or speed), but it can also be a change in direction.

* Acceleration is a vector in the same direction as the change in velocity. Since velocity is a vector, it can change in magnitude or in direction, or both. Acceleration is, therefore, a change in speed or direction, or both.

* **Instantaneous acceleration** (a), or acceleration at a specific instant in time, is obtained using the same process discussed for instantaneous velocity.

* The result is the derivative of the velocity function $v_{t}$, which is **instantaneous acceleration** and is expressed mathematically as:
> $a = \lim_{{\Delta t \to 0}} \frac{\Delta v}{\Delta t}$

or

$a_{t} = \frac{d}{dt}v_{t}$

# Position, Velocity, and Acceleration Relationships

* Position as a function of time:
> $x(t) = x_0 + v_0 t + \frac{1}{2} a t^2$

* Final Velocity $(v)$ as a function of initial velocity $(v_{0})$, acceleration $(a)$ and time $(t)$:
> $v = v_0 + a t$

* Position as a function of initial velocity $(v_{0})$, acceleration $(a)$ and time $(t)$:
> $x = x_0 + v_0 t + \frac{1}{2} a t^2$

* Final Velocity $(v)$ as a function of initial velocity $(v_{0})$, acceleration $(a)$ and displacement $(x)$:
> $v^2 = v_0^2 + 2 a (x - x_0)$

* Acceleration $(a)$ as as a function of initial velocity $(v_{0})$, final velocity $(v)$, and displacement $(x)$:
> $a = \frac{v^2 - v_0^2}{2(x - x_0)}$

```python
def PositionAsFunctionOfTime(a,t,x0=0,v0=0):
  return x0+v0*t+0.5*a*t**2
```

```python
def FinalVelocity(v0,a,t):
  return v0+a*t
```

```python
def FinalVelocitySquared(v0,a,x,x0=0):
  return v0**2+2*a*(x-x0)
```

```python
def Acceleration(v0,v,x,x0=0):
  return (v**2-v0**2)/(2*(x-x0))
```

**An airplane lands with an initial velocity of 70.0 m/s and then accelerates opposite to the motion at 1.50 m/s^2 for 40.0 s. What is its final velocity?**

```python
initialVelocity = 70.0 # in m/s
acceleration = -1.50 # negative because it's opposite to the motion
time = 40.0 # in seconds
```

```python
finalVelocity = FinalVelocity(initialVelocity,acceleration,time)
print(f"The final velocity of the airplane is {finalVelocity} m/s")
```
